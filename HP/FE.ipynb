```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

%matplotlib inline

dataset = pd.read_csv('Phising_Training_Dataset.csv')

data = dataset.drop("key", axis=1)
x = data.drop("Result", axis=1)
y = data["Result"]
y.replace(-1, 0, inplace=True)

print(data.shape, x.shape, y.shape)
# print(data.describe()) #data.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=0, stratify= y)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt

lda = LDA()
data_projected = lda.fit_transform(X_train, Y_train)
print(data_projected.shape)


markers = ['+','*']
colors = ['r','g']

fig = plt.figure(figsize=(12,12))
ax0 = fig.add_subplot(111)

for l,m,c in zip(np.unique(Y_train),markers,colors):
    ax0.scatter(data_projected[:,0][Y_train==l],data_projected[:,0][Y_train==l],c=c,marker=m)
```
